# Домашнее задание к лекции «Случайные события. Случайные величины»

Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш.

777: 200 руб.
999: 100 руб.
555: 50 руб.
333: 15 руб.
111: 10 руб.
*77: 5 руб.
**7: 3 руб.
*00: 2 руб.
**0: 1 руб.

Вопросы:

Выгодна ли игра игроку?
Сколько в среднем приобретает или теряет игрок за одну игру?

**Дополнительное задание повышенной сложности. **

Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования
– любая цифра

```python
# импорт библиотек
import numpy as np
```

```python
#  составим последовательность победных комбинаций

# 777: 200 руб.
# 999: 100 руб.
# 555: 50 руб.
# 333: 15 руб.
# 111: 10 руб.
# *77: 5 руб.
# **7: 3 руб.
# *00: 2 руб.
# **0: 1 руб.

wins = {999: 100, 777: 200, 555: 50, 333: 15, 111: 10, 0: 1} # словарь выигрышных комбинаций

count_wins = {100: 1, 200: 1, 50: 1, 15: 1, 10: 1, 0: 1}   # словарь кол-ва выигрышных исходов


for el in range(0,1000):
  if el not in wins:

    if el % 100 == 77:
      wins[el] = 5
      count_wins[5] = count_wins.get(5, 0) + 1

    elif el % 100 == 0:
      wins[el] = 2
      count_wins[2] = count_wins.get(2, 0) + 1

    elif el % 10 == 7:
      wins[el] = 3
      count_wins[3] = count_wins.get(3, 0) + 1

    elif el % 10 == 0:
      wins[el] = 1
      count_wins[1] = count_wins.get(1, 0) + 1

print(f'выигрышная комбинация: {wins}\nкол-во выигрышных исходов: {count_wins}')
```

```python
# Определим вероятность выигрыша

trials = 1_000_000    # кол-во попыток == сумма проигранных денег
victory = 0           # кол-во побед
win = 0               # сумма выигранных денег

for i in range(trials):
  random_score = np.random.randint(0, 1_000)
  if random_score in wins.keys():
    victory += 1
    win += wins[random_score]

p_monte_carlo = victory / trials    # Вероятность выпадения выигрышной комбинации

print(f'Кол-во побед: {victory}\nВероятность выпадения выигрышной комбинации: {p_monte_carlo}\n'
      f'Вероятность выигрыша {win/trials} на {trials} попыток')

if win - trials > 0:
  print(f'Игра выгодна, выигрыш: {win - trials}')
else:
  print(f'Игра не выгодна, проигрыш: {abs(win - trials)}')
```

```python
# Теоретически рассчитаем средний выигрыш (проигрыш) по формуле полной вероятности: P(A)=...

resault = 0

for key , val in count_wins.items():
  resault += key * val/1000

print(f'Теориетическая вероятность выигрыша: {resault}')

if trials * (resault - 1) > 0:
  print(f'Игра выгодна, выигрыш: {trials * (resault - 1) } ')
else:
  print(f'Игра не выгодна, проигрыш: {abs(trials * (resault - 1) ) } ')
```

```python
#  Отклонение эксперимент/теоретическое значение

p = abs( 1 - abs(win - trials) / abs (trials * (resault - 1) ) ) * 100

print(f'Отклонение экспериментального от теоретического значение не более {round(p,3)}%')
```
