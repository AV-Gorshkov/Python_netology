Определите, какому фильму было выставлено больше всего оценок 5.0.

```python
#  получение данных
param = r'/content/drive/My Drive/Нетология/Python_Pandas_Numpy/ДЗ_Дата_сет/'

file_1 = 'movies.csv'
file_2 = 'ratings.csv'

file_movies = pd.read_csv(f'{param}{file_1}', sep = ',')
file_rating = pd.read_csv(f'{param}{file_2}', sep = ',')

file_movies.head(10)
# file_rating.head(10)
```

```python
# Решение
# фильтруем датасет Rating по отметке "rating = 5.0" и считаем кол-во строк для каждого фильма MovieId.
# index[0] - фильм с наибольшим рейтингом
# находим название фильма в датасете Movies по movieId из датасета Rating

data_rating = file_rating[ (file_rating['rating']==5.0) ]['movieId'].value_counts().index[0]

data_movies = file_movies[ (file_movies['movieId']==data_rating) ]

data_movies.head()
```
