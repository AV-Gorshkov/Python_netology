Обязательная часть
Вам необходимо написать функцию, которая будет основана на поиске по сайту habr.com. Функция в качестве параметра должна принимать список запросов для поиска (например, ['python', 'анализ данных']) и на основе материалов, попавших в результаты поиска по каждому запросу, возвращать датафрейм вида:

<дата> - <заголовок> - <ссылка на материал>
В рамках задания предполагается работа только с одной (первой) страницей результатов поисковой выдачи для каждого запроса. Материалы в датафрейме не должны дублироваться, если они попадали в результаты поиска для нескольких запросов из списка.

Дополнительная часть (необязательная)
Функция из обязательной части задания должна быть расширена следующим образом:

кроме списка ключевых слов для поиска необходимо объявить параметр с количеством страниц поисковой выдачи. Т.е. при передаче в функцию аргумента 4 необходимо получить материалы с первых 4 страниц результатов;
в датафрейме должны быть столбцы с полным текстом найденных материалов и количеством лайков:
<дата> - <заголовок> - <ссылка на материал> - <текст материала> - <количество лайков>
ПРИМЕЧАНИЕ

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

```python
# Задание_1

url = 'https://habr.com/ru/search/'
list_post = ['python', 'анализ данных']

def get_post (query):

  df_post =pd.DataFrame()
  all_link = []

  for post in query:
    params = {  'page': 1, 'q': post }
    res = requests.get(url, params=params)
    time.sleep(0.3)
    soup = BeautifulSoup(res.text)
    articles = soup.find_all('div', class_='tm-article-snippet')

# собираем ссылки подходящие по списку из запроса
    for el in articles:

      link = el.find('a', class_='tm-title__link').get('href')
      all_link.append('https://habr.com'+link)

  all_link= set(all_link)

# формируем DataFrame
  for link in all_link:
    soup = BeautifulSoup(requests.get(link).text)

    try:
      title = soup.find('h1', 'tm-title tm-title_h1').text.strip()
      date =  pd.to_datetime(soup.find('time')['title']).date()
      row = {'date': date, 'title': title, 'link': link}
      df_post = pd.concat([ df_post, pd.DataFrame([row])] )
    except:
      pass

  return df_post.reset_index(drop=True)

res = get_post(list_post)
res
```

Дополнительная часть (необязательная)
Функция из обязательной части задания должна быть расширена следующим образом:

кроме списка ключевых слов для поиска необходимо объявить параметр с количеством страниц поисковой выдачи. Т.е. при передаче в функцию аргумента 4 необходимо получить материалы с первых 4 страниц результатов;
в датафрейме должны быть столбцы с полным текстом найденных материалов и количеством лайков:
<дата> - <заголовок> - <ссылка на материал> - <текст материала> - <количество лайков>

```python
# Задание_2
import re

url = 'https://habr.com/ru/search/'
list_post = ['python', 'анализ данных']
pat = r'Всего голосов\s(\d+)'

def get_post (query, pages=4):

  df_query = pd.DataFrame()
  all_link = []

  for post in query:
    for number in range(1, int(pages) + 1):

      params = {  'page': number, 'q': post }
      res = requests.get(url, params=params)
      time.sleep(0.3)
      soup = BeautifulSoup(res.text)
      articles = soup.find_all('div', class_='tm-article-snippet')

# собираем ссылки подходящие списку из запроса
      for el in articles:
        link = el.find('a', class_='tm-title__link').get('href')
        all_link.append('https://habr.com'+link)
  all_link= set(all_link)

# формируем DataFrame
  for link in all_link:
    soup = BeautifulSoup(requests.get(link).text)

    try:
      title = soup.find('h1', 'tm-title tm-title_h1').text.strip()
      date =  pd.to_datetime(soup.find('time')['title']).date()
      doc = soup.find('div', class_='tm-article-body').text.strip()
      try:
        rating = re.search( pat, str (soup.find('div', class_='tm-article-rating tm-data-icons__item').find('title').text.strip() ) ).group(1)
      except:
        rating = 0
      row = {'date': date, 'title': title, 'link': link, 'text': doc, 'rating': rating }
      df_query = pd.concat([ df_query, pd.DataFrame([row])] )
    except:
      pass

  return df_query.reset_index(drop=True)

res = get_post(list_post)
res.head(15)
```
